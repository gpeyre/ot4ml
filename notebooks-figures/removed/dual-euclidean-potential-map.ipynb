# Euclidean Dual Potential and Induced Brenier Map

For the quadratic cost, Kantorovich duality can be rewritten in terms of a convex potential $\phi$.  The optimal Monge map is then $T=\nabla\phi$.  This notebook draws a simple convex quadratic potential, its contour lines, and the displacement from red source points $x_i$ to blue target points $T(x_i)$.

```python
from pathlib import Path
import sys
import numpy as np
import matplotlib.pyplot as plt

ROOT = Path.cwd()
if not (ROOT / "notebooks-figures").exists():
    ROOT = ROOT.parent
sys.path.append(str(ROOT / "notebooks-figures"))

from figure_style import (
    RED, BLUE, VIOLET, ORANGE, GRAY, LIGHT_GRAY,
    DIRAC_MARKER_SIZE, setup_matplotlib, figure_dir, save_pdf,
    remove_axes, box_axes, padded_limits, interp_color, draw_point_clouds,
)

setup_matplotlib()

NAME = "dual-euclidean-potential-map"
OUT = figure_dir(NAME)
rng = np.random.default_rng(52)
```

We use an affine Brenier map $T(x)=Ax+b$ with $A$ symmetric positive definite, so that $T=\nabla\phi$ for $\phi(x)=\frac12 x^\top A x+b^\top x$.

```python
n = 42
x = rng.normal(size=(n, 2)) @ np.array([[0.62, 0.10], [0.00, 0.48]]) + np.array([-0.20, 0.05])
A = np.array([[1.28, 0.34], [0.34, 0.74]])
b = np.array([0.42, -0.18])
y = x @ A.T + b
pts = np.vstack([x, y])
(xlim, ylim) = padded_limits(pts, pad=0.32)
xx = np.linspace(xlim[0], xlim[1], 180)
yy = np.linspace(ylim[0], ylim[1], 180)
X, Y = np.meshgrid(xx, yy)
Z = 0.5 * (A[0,0]*X**2 + 2*A[0,1]*X*Y + A[1,1]*Y**2) + b[0]*X + b[1]*Y
```

The contour lines show the scalar potential.  The arrows are not arbitrary vector fields: they are normal to the level sets because they are gradients.

```python
fig, ax = plt.subplots(figsize=(2.55, 2.35))
ax.contour(X, Y, Z, levels=13, colors=[LIGHT_GRAY], linewidths=0.65, alpha=0.95)
for i in range(n):
    ax.annotate("", xy=y[i], xytext=x[i], arrowprops=dict(arrowstyle="->", color=VIOLET, lw=0.65, alpha=0.62, shrinkA=1.5, shrinkB=1.5), zorder=1)
ax.scatter(x[:,0], x[:,1], s=DIRAC_MARKER_SIZE, marker="o", color=RED, edgecolor="none", linewidth=0, zorder=3)
ax.scatter(y[:,0], y[:,1], s=DIRAC_MARKER_SIZE, marker="o", color=BLUE, edgecolor="none", linewidth=0, zorder=3)
ax.set_xlim(*xlim)
ax.set_ylim(*ylim)
ax.set_aspect("equal")
remove_axes(ax)
save_pdf(fig, OUT / "potential-map.pdf", pad_inches=0.06)
plt.close(fig)
```
